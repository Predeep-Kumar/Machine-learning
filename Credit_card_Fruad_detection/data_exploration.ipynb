```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# Load data

```python
data = pd.read_csv('Data/creditcard.csv')
```

# Basic statistics

```python
print(data.describe())
```

# Distribution of classes

```python

plt.figure(figsize=(8, 6))
sns.countplot(x='Class', data=data)
plt.title('Distribution of Classes')
plt.xlabel('Class (0: legitimate, 1: fraudulent)')
plt.ylabel('Count')
plt.show()
```

# Distribution of transaction amounts

```python
plt.figure(figsize=(10, 6))
sns.histplot(data['Amount'], bins=30, kde=True)
plt.title('Distribution of Transaction Amounts')
plt.xlabel('Amount')
plt.ylabel('Count')
plt.show()
```


# Time vs. Amount by class

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Time', y='Amount', hue='Class', data=data)
plt.title('Time vs. Amount by Class')
plt.xlabel('Time')
plt.ylabel('Amount')
plt.show()
```

# Correlation matrix

```python
plt.figure(figsize=(12, 8))
corr_features = ['Amount', 'V1', 'V2', 'V3', 'V4', 'V5', 'V6', 'V7', 'V8', 'V9', 'V10', 'Class']
corr = data[corr_features].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Correlation Matrix')
plt.show()
```

# Feature distributions by class (example with V1, V2, V3)

```python
features = ['V1', 'V2', 'V3', 'V4', 'V5']
for feature in features:
    plt.figure(figsize=(10, 6))
    sns.kdeplot(data.loc[data['Class'] == 0, feature], label='Legitimate', fill=True)
    sns.kdeplot(data.loc[data['Class'] == 1, feature], label='Fraudulent', fill=True)
    plt.title(f'Distribution of {feature} by Class')
    plt.xlabel(feature)
    plt.ylabel('Density')
    plt.legend()
    plt.show()
```
